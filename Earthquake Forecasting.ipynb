```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm
import matplotlib
df = pd.read_csv('desktop/database.csv')
```

```python
df.dropna(subset=['Magnitude','Date', 'Latitude', 'Longitude'])
X = df[['Magnitude', 'Date', 'Latitude', 'Longitude']]
X
```

```python
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
plt.style.use("ggplot")
X["Magnitude"].plot(color="purple")
plt.show()
```

```python
K = df[['Latitude', 'Longitude']]
from sklearn.cluster import KMeans
K_Means = KMeans(n_clusters=2).fit(K)
K.plot.scatter(x='Latitude', y='Longitude', c='DarkBlue')
K_Means.predict(K)
K_clustering = K_Means.cluster_centers_
for i in range (0,2):
    plt.plot(K_clustering[i][0], K_clustering[i][1], 'ro')
plt.show()
```

```python
distances = []
for i in range (1,16):
    K_Means = KMeans(n_clusters=i).fit(K)
    distances.append(K_Means.inertia_)
for i in range (0,15):
    plt.plot(i+1, distances[i], 'ro')
plt.show()
```

```python
dates = []
magnitudes = []
count = 0
for i in range(0,23411):
    if (abs(-40-X.loc[i].at["Latitude"]) <= 6) and (abs(X.loc[i].at["Longitude"]-172) <= 6):
        dates.append(X.loc[i].at["Date"])
        magnitudes.append(X.loc[i].at["Magnitude"])
        count += 1
count
```

```python
from fbprophet import Prophet
from fbprophet.plot import plot_plotly
from sklearn.model_selection import train_test_split
y = magnitudes
X = pd.DataFrame()
X['ds'] = pd.to_datetime(dates, utc=True)
X['ds'] = X['ds'].dt.tz_localize(None)
X['y'] = y
prediction = Prophet()
prediction.fit(X)
future = prediction.make_future_dataframe(periods=240, freq='m')
forecast = prediction.predict(future)
figure = prediction.plot(forecast)
figure = prediction.plot_components(forecast)
```
